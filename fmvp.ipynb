```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('fmvp.csv')
ppgstats = (np.mean(data['PPG']), np.std(data['PPG']))
apgstats = (np.mean(data['APG']), np.std(data['APG']))
rpgstats = (np.mean(data['RPG']), np.std(data['RPG']))
print(ppgstats)
print(apgstats)
print(rpgstats)
```

```python




# Histogram for Points Per Game (PPG)
plt.hist(data['PPG'], bins=10, alpha=0.7, edgecolor='black')
plt.title('Histogram of PPG')
plt.xlabel('PPG')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

```

```python
plt.hist(data['APG'], bins=10, alpha=0.7, edgecolor='black')
plt.title('Histogram of APG')
plt.xlabel('APG')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.hist(data['RPG'], bins=10, alpha=0.7, edgecolor='black')
plt.title('Histogram of RPG')
plt.xlabel('RPG')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
box = plt.boxplot([data['FG%'], data['PPG'], data['RPG'], data['APG'], data['BLKPG']], labels=['FG%', 'PPG', 'RPG', 'APG', 'BLKPG'], patch_artist=True)

colors = ['pink', 'lightblue', 'green', 'lavender', 'yellow']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)


plt.title('Box Plot of FG%, PPG, RPG, APG, BLKPG')
plt.ylabel('Values')
plt.show()
```
